```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('dqn_conn0_ep1.csv')

# Extract relevant columns for plotting
time_steps = data['step']
total_waiting_time = data['system_total_waiting_time']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time_steps, total_waiting_time, label='Total Waiting Time', linewidth=1)
plt.xlabel('Time Step (s)')
plt.ylabel('Total Waiting Time of Vehicles (s)')
plt.title('Total Waiting Time of Vehicles Over Time')
plt.grid(True)
plt.show()
```

```python
# Plot: System Total Stopped Over Time
plt.figure(figsize=(10, 6))
plt.plot(data['step'], data['system_total_stopped'], label='Total Stopped Vehicles', color='orange')
plt.xlabel('Time Step (s)')
plt.ylabel('Total Stopped Vehicles')
plt.title('Total Stopped Vehicles Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot: Mean Waiting Time Over Time
plt.figure(figsize=(10, 6))
plt.plot(data['step'], data['system_mean_waiting_time'], label='Mean Waiting Time', color='green')
plt.xlabel('Time Step (s)')
plt.ylabel('Mean Waiting Time (s)')
plt.title('Mean Waiting Time of Vehicles Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot: Agents Total Stopped Over Time
plt.figure(figsize=(10, 6))
plt.plot(data['step'], data['agents_total_stopped'], label='Agents Total Stopped', color='purple')
plt.xlabel('Time Step (s)')
plt.ylabel('Total Stopped Agents')
plt.title('Agents Total Stopped Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot: Agents Total Accumulated Waiting Time
plt.figure(figsize=(10, 6))
plt.plot(data['step'], data['agents_total_accumulated_waiting_time'], label='Accumulated Waiting Time', color='brown')
plt.xlabel('Time Step (s)')
plt.ylabel('Accumulated Waiting Time (s)')
plt.title('Agents Total Accumulated Waiting Time Over Time')
plt.grid(True)
plt.legend()
plt.show()
```
